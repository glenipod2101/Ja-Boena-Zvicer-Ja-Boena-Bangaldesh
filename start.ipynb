```python
real_madrid_matches_home = df[(df['AwayTeam'] == 'Barcelona')]
```

```python
# 1XBA = AWAY TO WIN
# 1XBH = HOME TO WIN


# PER MOS TA HANGER SHUM ME EKIPE T FORTA SI PSH BARCELONA SHEF CDO KOEF POSHT 4, POR MUND TA NDROSH

real_madrid_low_odds = real_madrid_matches_home[real_madrid_matches_home['1XBA'] < 4.0]
```

```python
# per ti ber gjerat me t kollajshme

# perdorim vetem Date Time ,Kush Ka Lujt, dhe sa ka dal me gjith koef e neshjes qe t fitoj skuadra e shpis
final_df = real_madrid_low_odds[['Date', 'Time', 'HomeTeam', 'AwayTeam', '1XBA', 'FTR']]
```

```python
# tashi shtojm i kollon qe multiplies 100 * koef e asaj dite
# ke hapi 2, nese FTR esht H, pra home team ka fitu, shtojm profit = 100* 1XBH
# nese FTR esht A dhe dikush e ka qi real madridin n byth n fush t vet ne humim 100 ero

def calculate_profit(row):
    if row['FTR'] == 'A':  # Fito Reali Modh
        return ((100 * row['1XBA']) - 100)  # Profit = (100 * home win odds) - 100
    elif row['FTR'] == 'H':  # Hum Reali Modh
        return -100  # Lodren Mrena
    else:  # Draw (FTR == 'D')
        return -100  # Lodren Prap Mrena
```

```python
final_df['Profit'] = final_df.apply(calculate_profit, axis=1)
```
